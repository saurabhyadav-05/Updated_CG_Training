```python
# Scenario: Sports Event Analysis
# Context:
# You are part of a sports analytics team working for a global media company. Your manager has asked you to analyze match reports and press releases
# to identify key entities such as teams, players, locations, dates, and numbers. This information will help the company build automated summaries and
# highlight reels for fans.
# Task:
# You receive the following match report excerpt:
# "Manchester United defeated Real Madrid 3-2 at Wembley Stadium, London. Cristiano Ronaldo scored twice, while Marcus Rashford netted the winning goal in July 2024"
# Your job is to run Named Entity Recognition (NER) using spacy to automatically extract structured insights.
# 14:44
# Type a message

import spacy

# Load the English model
nlp = spacy.load("en_core_web_sm")

# Match report text
text = """Manchester United defeated Real Madrid 3-2 at Wembley Stadium, London.
Cristiano Ronaldo scored twice, while Marcus Rashford netted the winning goal in July 2024"""

# Process the text
doc = nlp(text)

# Extract entities
entities = []
for ent in doc.ents:
    entities.append((ent.text, ent.label_))

# Return entities
entities
```

```python
# 📘 Scenario: Smart Manufacturing Insights
# Context:
# You are part of a data science team in a smart manufacturing company. Your manager wants you to explore how different
# technical concepts (like maintenance, telemetry, quality control, supply chain) are related in company documents.
# By training a small Word2Vec model, you can uncover semantic similarities between words and measure how closely they are
#  connected. This helps in building recommendation systems, knowledge graphs, or even automated reporting tools.
# Task:
# Given a short corpus describing manufacturing operations, train a Word2Vec model and analyze word similarities.


# Word embeddings with gensim Word2Vec
# Setup: python -m pip install gensim nltk

# Smart Manufacturing Insights using Word2Vec

# Smart Manufacturing Insights using Word2Vec

# Install required libraries
!pip install gensim nltk

# Import libraries
import nltk
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec

# Download required tokenizer resources
nltk.download('punkt')
nltk.download('punkt_tab')

# Manufacturing corpus
corpus = [
    "The maintenance team monitors machine health using telemetry data",
    "Predictive maintenance helps reduce machine downtime",
    "Quality control ensures that manufactured products meet standards",
    "Supply chain management tracks materials and logistics",
    "Telemetry sensors collect temperature and vibration data",
    "Maintenance engineers analyze machine performance",
    "Quality inspection improves product reliability",
    "Supply chain disruptions affect manufacturing operations"
]

# Tokenize sentences
tokenized_corpus = []
for sentence in corpus:
    tokens = word_tokenize(sentence.lower())
    tokenized_corpus.append(tokens)

# Train Word2Vec model
model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)

# Find similar words
print("Words similar to 'maintenance':")
print(model.wv.most_similar("maintenance"))

# Word similarity
print("\nSimilarity between 'maintenance' and 'machine':")
print(model.wv.similarity("maintenance", "machine"))

# Word vector
print("\nVector for 'maintenance':")
print(model.wv["maintenance"])
```

```python
# 📘 Scenario: Healthcare Research Insights
# Context:
# You are part of a data science team in a healthcare research institute. Your manager wants you to analyze medical reports
# and research papers to explore how different technical concepts (like diagnosis, treatment, symptoms, patients, medication) are related.
# By training a small Word2Vec model, you can uncover semantic similarities between words and measure how closely they are connected.
# This helps in building clinical decision support systems, medical knowledge graphs, or even automated literature reviews.
# Task:
# Given a short corpus describing healthcare operations, train a Word2Vec model and analyze word similarities.

# 📘 Healthcare Research Insights using Word2Vec
!pip install gensim nltk

import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from gensim.models import Word2Vec

# Download required NLTK resources
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize tokenizer, stopwords, and lemmatizer
tokenizer = RegexpTokenizer(r'\w+')
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Healthcare corpus
corpus = [
    "Doctors diagnose diseases using medical tests and patient symptoms",
    "Treatment plans include medication and therapy for patients",
    "Medical researchers analyze patient data to improve diagnosis",
    "Symptoms such as fever and cough indicate possible infection",
    "Medication helps treat chronic diseases and improve patient recovery",
    "Hospitals maintain electronic health records for patients",
    "Clinical trials test new treatments and medications",
    "Doctors monitor patient symptoms to adjust treatment plans"
]

# Tokenize and preprocess corpus
tokenized_corpus = []
for sentence in corpus:
    tokens = tokenizer.tokenize(sentence.lower())
    cleaned_tokens = [
        lemmatizer.lemmatize(word) for word in tokens
        if word not in stop_words
    ]
    tokenized_corpus.append(cleaned_tokens)

# Train Word2Vec model
model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)

# Find words similar to "treatment"
print("Words similar to 'treatment':")
print(model.wv.most_similar("treatment"))

# Similarity between "diagnosis" and "treatment"
print("\nSimilarity between 'diagnosis' and 'treatment':")
print(model.wv.similarity("diagnosis", "treatment"))

# Word vector for "patient"
print("\nWord vector for 'patient':")
print(model.wv["patient"])
```
